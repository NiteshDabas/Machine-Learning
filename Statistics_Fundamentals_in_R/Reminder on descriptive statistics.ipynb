# Reminder on descriptive statistics

## Histogram and boxplot

```r
data(iris)
SepalLength = iris$Sepal.Length
```

With Histogram we can manipulate data to tell a story WE want. With boxplots, we cannot tweak data.

```r
par(mfrow=c(2,2))
hist(SepalLength)
hist(SepalLength, breaks=50) #by changing breaks we can make a new story
hist(SepalLength, breaks=2)
boxplot(SepalLength,horizontal = TRUE) #no other parameter to tune
```

Comparision between Histogram and Boxplot

```r
hist(SepalLength,freq=FALSE, col=4)
lines(density(SepalLength),lwd=2,col=2)
par(new=TRUE)
boxplot(SepalLength,horizontal = TRUE, col=7)
```

Boxplot we can draw for Multivariate data in same frame while istogram gives different plot for diff data.

```r
par(mfrow=c(2,3))
boxplot(iris[,-5], freq=FALSE)
for (i in 1:4) hist(iris[,i])
```

## Scatter plots and Pair plots

```r
#Scatter Plot
cls = as.numeric(iris$Species)
plot(iris[,1:2], type='p', col=cls, pch=cls)
legend("bottomright",legend = c("Versicolor","Setosa","Virginica"),col=1:3,pch=1:3)
```

```r
#Pair Plot with legend
pairs(iris[,-5], col=as.numeric(iris$Species), oma=c(3,3,3,15))
par(xpd = TRUE)
legend("bottomright", fill=unique(iris$Species), legend=c("Versicolor", "Sentosa", "Virginica"), col=1:3, pch=19)
```

## About the multivariate numerical indicators 

```r
#Mean of a single column
Mean = mean(iris$Sepal.Length)
Mean

#Mean of seperate columns in a matrix
mu = colMeans(iris[,-5])
mu

#Covariance Matrix
CovMatrix = var(iris[,-5])
CovMatrix

#Correlation Matrix
CorrMatrix = cor(iris[,-5])
CorrMatrix
```
